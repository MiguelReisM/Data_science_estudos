```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../data/gas_prices.csv", encoding="latin1")
```

```python
plt.title('Price in US$')

list_country = ['Australia', 'USA', 'Italy', 'Mexico', 'France', 'Germany']
for c in df:
    if c in list_country:
        plt.plot(df['Year'], df[c], label = c, marker = '.')

plt.xticks(df['Year'][::2]) #le todos os valores de anos, mas filtra e mostra de 2 em 2 anos

plt.xlabel('Years')
plt.ylabel('US$')

plt.legend() #mostra a legenda
plt.savefig("../img/country.png")
plt.savefig("../img/country.pdf")
plt.savefig("../img/country.jpg")
plt.show()
```
