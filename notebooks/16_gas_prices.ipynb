```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../data/gas_prices.csv", encoding="latin1")
```

```python
plt.title('Price in US$')

plt.plot(df['Year'], df['Australia'], 'r.-' ,label = 'Australia')
plt.plot(df['Year'], df['Italy'], 'y.-' ,label = 'Italy')
plt.plot(df['Year'], df['USA'], 'b.-' ,label = 'USA')
plt.plot(df['Year'], df['Japan'], 'g.-' ,label = 'Japan')

plt.xticks(df['Year'][::2]) #le todos os valores de anos, mas filtra e mostra de 2 em 2 anos

plt.xlabel('Years')
plt.ylabel('US$')

plt.legend() #mostra a legenda adicionada pelo label
plt.show()
```
