```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("../data/gas_prices.csv", encoding="latin1")
```

```python
plt.title('Valores do combustivel em US$')

plt.plot(df['Year'], df['Australia'], label = 'Australia')
plt.plot(df['Year'], df['Italy'], label = 'Italy')
plt.plot(df['Year'], df['USA'], label = 'USA')
plt.plot(df['Year'], df['Japan'], label = 'Japan')

plt.xticks(df['Year'][::2]) #le todos os valores de anos, mas filtra e mostra de 2 em 2 anos
plt.legend() #mostra a legenda adicionada pelo label
plt.show()
```
